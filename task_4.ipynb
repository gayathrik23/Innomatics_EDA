# Univariate Analysis

```python
# Find outliers using the IQR method for each numerical column
def find_outliers_iqr(data):
    outliers = {}
    for column in data.select_dtypes(include=['int64', 'float64']):
        Q1, Q3 = data[column].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = data[(data[column] < lower_bound) | (data[column] > upper_bound)][column]
    return outliers

# Call the function to find outliers
outliers = find_outliers_iqr(data)

# Print outliers for each numerical column
for column, values in outliers.items():
    print(f"Outliers in column '{column}':")
    print(values)
    print()
```

The existence of outliers underscores the importance of comprehensive data exploration, analytical scrutiny, and well-informed decision-making to uphold the credibility and robustness of subsequent analyses or modeling endeavors.

```python
#2
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(data['Gender'], data['Specialization'])

chi2, p_value, _, _ = chi2_contingency(contingency_table)


print(p_value)
```

The p-value indicates the probability of observing the data if the two categorical variables (gender and specialization) were independent.

A low p-value suggests a significant association between gender and specialization, while a high p-value suggests independence or lack of association.
